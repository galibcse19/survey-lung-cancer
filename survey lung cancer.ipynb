```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('survey lung cancer_clear.csv')
```

```python
df.head()
```

```python
%matplotlib inline
plt.title('survey lung cancer analysis')
plt.xlabel('AGE')
plt.ylabel('LUNG_CANCER')
plt.scatter(df.AGE,df.LUNG_CANCER,color='red',marker='+')
```

```python
x=df.drop(['LUNG_CANCER'],axis='columns')
```

```python
x.head()
```

```python
y=df['LUNG_CANCER']
```

```python
y.head()
```

```python
df.isnull().sum()
```

# Data preprocessing  by using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data_encode=LabelEncoder()
```

```python
x['GENDER_EN'] = data_encode.fit_transform(x['GENDER'])
```

```python
x
```

```python
x=x.drop(['GENDER'],axis='columns')
```

```python
x
```

# Data spliting in Train, Test----

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.25,random_state=1)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
ytrain.head()
```

```python
ytest.head()
```

# SVM

```python
from sklearn.svm import SVC
```

```python
model_svm=SVC()
```

```python
model_svm.fit(xtrain,ytrain)
```

```python
model_svm.score(xtest,ytest)
```

```python
model_svm.predict([[69,1,2,2,1,1,2,1,2,2,2,2,2,2,1]])
```

# Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dt=DecisionTreeClassifier()
```

```python
model_dt.fit(xtrain,ytrain)
```

```python
model_dt.score(xtest,ytest)
```

# Random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_RF=RandomForestClassifier()
```

```python
model_RF.fit(xtrain,ytrain)
```

```python
model_RF.score(xtest,ytest)
```
